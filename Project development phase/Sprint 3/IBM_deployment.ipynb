```python
#importing the necessory libraries

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#define the parameters/arguments for imagedatagenarator class
train_datagen=ImageDataGenerator(rescale=1./255,
                                 shear_range=0.2,
                                 rotation_range=180,
                                 zoom_range=0.2,
                                 horizontal_flip=True)
```

```python

test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
#applying imagedatagenerator functionality to trainset 
x_train=train_datagen.flow_from_directory('/content/Dataset/Dataset/train_set',
                                          target_size=(128,128),
                                          batch_size=32,
                                          class_mode='binary')
```

```python
#applying imagedatagenerator functionality to testset
x_test=test_datagen.flow_from_directory('/content/Dataset/Dataset/test_set',
                                        target_size=(128,128),
                                        batch_size=32,
                                        class_mode='binary')
                                    
```

```python
#Importing the model libraries

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolution2D, MaxPooling2D, Flatten, Dense
```

```python
#initializing the model
model=Sequential()
```

```python
#adding CNN layers
model.add(Convolution2D(32,(3,3),input_shape=(128,128,3),activation='relu'))    # convolution layer
model.add(MaxPooling2D(pool_size=(2,2)))                                        #maxpool layer
model.add(Flatten())                                                            #flatten layer
```

```python
#adding dense layers
model.add(Dense(150,activation='relu'))                                         #hidden layer
model.add(Dense(150,activation='relu'))                                         #hidden layer

model.add(Dense(1,activation="sigmoid"))                                        #output layer

```

```python
#configuring the learning process
model.compile(loss='binary_crossentropy',optimizer='adam',
              metrics=['accuracy'])

```

```python
#training the model
model.fit_generator(x_train,
                    steps_per_epoch=14,
                    epochs=10,
                    validation_data=x_test,
                    validation_steps=4)
                  
```

```python
#saving the model
model.save("forest.h5")

!tar -zcvf forest-fire-detection.tgz forest.h5


```
